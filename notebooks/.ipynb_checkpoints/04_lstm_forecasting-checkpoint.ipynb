```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
df = pd.read_csv("../data/sales_data.csv")

df['data'] = pd.to_datetime(df['data'])
df.set_index('data', inplace=True)
```

```python
sales = df[['venda']]
```

```python
scaler = MinMaxScaler(feature_range=(0,1))    #scale the data 
scaled_sales = scaler.fit_transform(sales)
```

```python
X, y = [], []           #create sequence (window=12)

for i in range(12, len(scaled_sales)):
    X.append(scaled_sales[i-12:i])
    y.append(scaled_sales[i])

X, y = np.array(X), np.array(y)
```

```python
X = X.reshape((X.shape[0], X.shape[1], 1))   #reshape for LSTM
```

```python
model = Sequential()       #LSTM build
model.add(LSTM(50, activation='relu', input_shape=(X.shape[1], 1)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=20, batch_size=16, verbose=1)
```

```python
pred_scaled = model.predict(X[-12:])   #Generate predictions (LAST 12 PERIODS)
pred = scaler.inverse_transform(pred_scaled)
```

```python
plt.figure(figsize=(12,5))           #Plot Actual vs LSTM Forecast
plt.plot(df.index[-12:], df['venda'][-12:], label="Actual Sales")
plt.plot(df.index[-12:], pred.flatten(), label="LSTM Forecast")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("LSTM Sales Forecast")
plt.legend()
plt.show()
```

The LSTM model captures long-term dependencies in historical sales data and performs well in modeling non-linear demand patterns. The forecast closely follows recent sales behavior.
