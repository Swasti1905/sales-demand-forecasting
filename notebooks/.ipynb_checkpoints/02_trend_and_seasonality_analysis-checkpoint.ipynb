```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("../data/sales_data.csv")
```

```python
df['data'] = pd.to_datetime(df['data'])
df.set_index('data', inplace=True)
```

```python
df.head()
```

df['venda'].plot(figsize=(12,5), title="Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()


```python
df['venda'].isna().sum()  #handle missing values
```

```python
result = seasonal_decompose(df['venda'], model='additive', period=12)      #trend and seasonality decomposition #it will show 1. Observed  Original sales 2.Trend  Long-term movement 3. Seasonal  Repeating pattern 4. Residual  Noise
result.plot()
plt.show()
```

Trend & Seasonality Insights
The trend component shows the overall long-term movement in sales.
The seasonal component indicates repeating demand patterns over time.
Residuals capture random fluctuations not explained by trend or seasonality.
