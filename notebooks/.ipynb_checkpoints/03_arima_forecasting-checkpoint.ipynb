```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
df = pd.read_csv("../data/sales_data.csv")
```

```python
df['data'] = pd.to_datetime(df['data'])
df.set_index('data', inplace=True)
```

```python
sales = df['venda']
```

```python
train = sales.iloc[:-12]
test = sales.iloc[-12:]
```

```python
df.index.to_series().diff().value_counts().head()
```

```python
df = df.asfreq('D')
```

```python
df = df.asfreq('MS')
```

```python
df = df.asfreq('W')
```

```python
df['venda'] = df['venda'].ffill()
```

```python
sales = df['venda']            #ARIMA
train = sales.iloc[:-12]
test = sales.iloc[-12:]

model = ARIMA(train, order=(5,1,0))
model_fit = model.fit()

forecast = model_fit.forecast(steps=12)
```

```python
plt.figure(figsize=(12,5))
plt.plot(test.index, test, label="Actual Sales")
plt.plot(test.index, forecast, label="ARIMA Forecast")
plt.legend()
plt.show()
```

I ensured the datetime index had proper frequency information so that the ARIMA model could generate time-aware forecasts.
